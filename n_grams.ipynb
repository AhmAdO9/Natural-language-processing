## N-grams:

```python
import pandas as pd
import os
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
import re



df = pd.read_csv('spam_collection/SMSSpamCollection', 
                 sep='\t', 
                 names=['label', 'message'])

lemmatizer = WordNetLemmatizer()

lemma = lemmatizer.lemmatize

stopwords = stopwords.words('english')

lemmatized_docs = []

for i in range(len(df['message'])):
    words = re.sub('[^a-zA-Z ]', '', df['message'][i])
    words = words.split(' ')
    lemmatized = [lemma(word.lower()) for word in words]
    lemmatized_docs.append(' '.join(lemmatized))
```

```python
countVector = CountVectorizer(max_features=200, 
                              binary=True, 
                              ngram_range=(2,2))
```

```python
X = countVector.fit_transform(lemmatized_docs)
```

```python
X.toarray()
```
